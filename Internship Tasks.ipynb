```python
#Task 2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

#print(df.head())
print(df.isnull().sum())

df['Age'].fillna(df['Age'].median(),inplace = True)
df['Embarked'].fillna(df['Embarked'].mode()[0] , inplace=True)
df['Cabin'].fillna('Unknown' , inplace=True)
df['Fare'].fillna(df['Fare'].median() , inplace = True)

print(df.isnull().sum())

df.drop(['PassengerId', 'Ticket', 'Name'], axis=1, inplace=True)

print(df.describe())
print(df.describe(include=['O']))

#Histogram for Gender
plt.figure(figsize=(10,6))
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Frequency')
sns.histplot(df['Sex'], bins=30 , kde= True)


#histogram for Age 
plt.figure(figsize=(10, 6))
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
sns.histplot(df['Age'], bins=30, kde=True)

sns.pairplot(df[['Age' ,'Fare' , 'SibSp' , 'Parch' , 'Survived']])

#survival with respect to passenger class
plt.figure(figsize = (8,6))
sns.barplot(data=df , x='Pclass' , y='Survived')
plt.title('Survival v.r.t Passenger class')
plt.xlabel("Pclass")
plt.ylabel('Survived')
plt.show

#survival with respect to gender
plt.figure(figsize = (8,6))
sns.barplot(data=df , x='Sex' , y='Survived')
plt.title('Survival v.r.t gender')
plt.xlabel("Gender")
plt.ylabel('Survived')
plt.show()

#survival by fare
plt.figure(figsize = (8,6))
sns.histplot(df[df['Survived'] == 1]['Fare'] , bins=30 , kde = True , color='green' , label = 'Survived')
sns.histplot(df[df['Survived'] == 0]['Fare'] , bins=30 , kde = True , color='red' , label = 'Not Survived')
plt.title("Survival Rate by Fare")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

#Survival by age 
plt.figure(figsize = (8,6))
sns.histplot(df[df['Survived'] == 1]['Age'] , bins=30 , kde = True , color='green' , label = 'Survived')
sns.histplot(df[df['Survived'] == 0]['Age'] , bins=30 , kde = True , color='red' , label = 'Not Survived')
plt.title("Survival Rate by age")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Task-03
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

df = pd.read_csv('diabetes.csv')

print(df.head())
print(df.columns)

feature_cols = ['Pregnancies', 'Insulin', 'BMI', 'Age', 'Glucose', 'BloodPressure', 'DiabetesPedigreeFunction']
target_col = 'Outcome'

x = df[feature_cols]
y = df[target_col]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)


dtc = DecisionTreeClassifier()
dtc = dtc.fit(x_train, y_train)

y_pred = dtc.predict(x_test)

accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Task-04
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

df = pd.read_csv('twitter_training.csv')

print(df.head())
print(df.columns)
sid = SentimentIntensityAnalyzer()

def get_sentiment(text):
    sentiment = sid.polarity_scores(text)
    return sentiment
    
text_column = 'Positive'  

df['sentiment'] = df[text_column].apply(get_sentiment)
df['compound'] = df['sentiment'].apply(lambda x: x['compound'])
df['sentiment_class'] = df['compound'].apply(lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral'))

print(df.head())

sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='sentiment_class', order=['positive', 'neutral', 'negative'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
#task-05

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('UK_Accident.csv')
df.head(1)

df = df.dropna(subset = ['Date' , 'Time' , 'Weather_Conditions' , 'Road_Surface_Conditions' , 'Light_Conditions','Accident_Severity'])

df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M').dt.time
df['Hour'] = df['Time'].apply(lambda x: x.hour)

# accident with respect to Weather Condtions
plt.figure(figsize = (12,6))
weather_order= df['Weather_Conditions'].value_counts().iloc[:10].index
sns.countplot(data=df , y="Weather_Conditions" , order=weather_order)
plt.title("accident with respect to Weather Condtions")
plt.xlabel("Number of accidents")
plt.ylabel("Weather conditions")
plt.show()

#Accident with respect to road surface conditions
plt.figure(figsize = (12,6))
road_order= df['Road_Surface_Conditions'].value_counts().iloc[:10].index
sns.countplot(data=df , y="Road_Surface_Conditions" , order=road_order)
plt.title("#Accident with respect to road surface conditions")
plt.xlabel("Number of accidents")
plt.ylabel("Weather conditions")
plt.show()

#Accident with respect to hour of the day 
plt.figure(figsize = (12,6))
sns.boxplot(data=df , x='Hour' , y='Accident_Severity')
plt.title("#Accident with respect time of the day")
plt.xlabel("Accident Severity")
plt.ylabel("Time of the Day")
plt.show()

#severity of accident with respect to weather
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Weather_Conditions', y='Accident_Severity')
plt.xticks(rotation=90)
plt.title('Accident Severity by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Accident Severity')
plt.show()

```
